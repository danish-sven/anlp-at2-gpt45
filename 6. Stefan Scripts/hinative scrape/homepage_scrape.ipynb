```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import StaleElementReferenceException
import time

# Replace with your Google email and password
email = "stefan.hall@student.uts.edu.au"
user_password = "this_is_my_password"

webdriver_path = 'chromedriver_mac64/chromedriver'
driver = webdriver.Chrome(executable_path=webdriver_path)

driver.get("https://hinative.com/")


def scroll_to_bottom():
    """
    Scroll to the bottom of the webpage using JavaScript.
    Wait for 3 seconds for the content to load, and continue scrolling until no new content is loaded.
    """
    previous_height = driver.execute_script("return document.body.scrollHeight")

    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(3)

        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == previous_height:
            break
        previous_height = new_height


try:
    # Log in
    login_button = driver.find_element(By.CSS_SELECTOR, ".signin_manual_anchor")
    login_button.click()

    username = driver.find_element(By.ID, "user_login")
    username.send_keys(email)

    password = driver.find_element(By.ID, "user_password")
    password.send_keys(user_password)

    submit_button = driver.find_element(By.XPATH, '//input[@type="submit"]')
    submit_button.click()

    # Wait for questions to load
    WebDriverWait(driver, 30).until(
        EC.presence_of_element_located((By.CSS_SELECTOR, ".d_block"))
    )

    # Initialize an empty set to store unique question URLs
    question_urls = set()

    # Read the existing URLs from the file, if the file exists
    try:
        with open("urls.txt", "r") as file:
            for line in file:
                question_urls.add(line.strip())
    except FileNotFoundError:
        pass

    # Continuously scrape URLs while scrolling
    while True:
        # Get all question blocks on the page
        question_blocks = driver.find_elements_by_css_selector(".d_block")

        for block in question_blocks:
            try:
                # Extract the question URL
                question_url = block.find_element(By.XPATH, ".//div[@class='container_fukidashi']/a").get_attribute("href")
                # Extract the question text
                question_text = block.find_element(By.CSS_SELECTOR, ".mod_question_content_decorated").text

                # Check if the URL is not already in the set and the question text contains the desired phrase
                if question_url not in question_urls and "Does this sound natural?" in question_text:
                    badge_items = block.find_elements(By.CSS_SELECTOR, ".badge_item")
                    for badge_item in badge_items:

                        # Check if the badge item has a number and if it's greater than or equal to 1
                        if badge_item.text.isdigit() and int(badge_item.text) >= 1:
                            # Add the URL to the set
                            question_urls.add(question_url)
                            print(question_url)

                            # Write the URL to the file
                            with open("urls.txt", "a") as file:
                                file.write(question_url + "\n")
                            break
            except StaleElementReferenceException:
                continue

        # Scroll to the bottom of the page to load more questions
        scroll_to_bottom()

finally:
    time.sleep(30)
    driver.quit()
```
